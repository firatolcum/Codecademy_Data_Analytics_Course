### Side by Side Bars <a class="anchor" id="2"></a>
[Table of Contents](#0.1)

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
sales1 =  [91, 76, 56, 66, 52, 27]
sales2 = [65, 82, 36, 68, 38, 40]
```

```python
n = 1  # This is our first dataset (out of 2)
t = 2 # Number of dataset
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store1_x = [t*element + w*n for element
             in range(d)]
store1_x
```

```python
n = 2  # This is our second dataset (out of 2)
t = 2 # Number of datasets
d = 6 # Number of sets of bars
w = 0.8 # Width of each bar
store2_x = [t*element + w*n for element
             in range(d)]
store2_x
```

```python
plt.bar(store1_x, sales1)
plt.bar(store2_x, sales2)

plt.show()
```

### Error Bars <a class="anchor" id="4"></a>
[Table of Contents](#0.1)

To display error visually in a bar chart, we often use error bars to show where each bar could be, taking errors into account.

Each of the black lines is called an error bar. The taller the bar is, the more uncertain we are about the height of the blue bar. The horizontal lines at the top and bottom are called caps. They make it easier to read the error bars.

If we wanted to show an error of +/- 2, we would add the keyword `yerr=2` to our `plt.bar` command. To make the caps wide and easy to read, we would add the keyword `capsize=10`:

```python
values = [10, 13, 11, 15, 20]
yerr = 2

plt.bar(range(len(values)), values, yerr=yerr, capsize=10)
plt.show()
```

```python
values = [10, 13, 11, 15, 20]
yerr = [1, 3, 0.5, 2, 4]

plt.bar(range(len(values)), values, yerr=yerr, capsize=10)
plt.show()
```

```python
drinks = ["cappuccino", "latte", "chai", "americano", "mocha", "espresso"]
ounces_of_milk = [6, 9, 4, 0, 9, 0]
error = [0.6, 0.9, 0.4, 0, 0.9, 0]
```

```python
plt.bar(range(len(ounces_of_milk)), ounces_of_milk, yerr = error, capsize = 5)

plt.show()
```

### Fill Between <a class="anchor" id="5"></a>
[Table of Contents](#0.1)

- In Matplotlib, we can use `plt.fill_between()` to shade error. This function takes three arguments:
    - x-values  this works just like the x-values of `plt.plot()`
    - lower-bound for y-values  sets the bottom of the shaded area
    - upper-bound for y-values  sets the top of the shaded area

- Generally, we use `.fill_between()` to create a shaded error region, and then plot the actual line over it. We can set the alpha keyword to a value between 0 and 1 in the `.fill_between()` call for transparency so that we can see the line underneath.

```python
x_values = range(10)
y_values = [10, 12, 13, 13, 15, 19, 20, 22, 23, 29]
y_lower = [i - 2 for i in y_values]
y_upper = [i + 2 for i in y_values]
 
plt.fill_between(x_values, y_lower, y_upper, alpha=0.2) #this is the shaded error
plt.plot(x_values, y_values) #this is the line itself
plt.show()
```

```python
months = range(12)
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
revenue = [16000, 14000, 17500, 19500, 21500, 21500, 22000, 23000, 20000, 19500, 18000, 16500]
```

```python
ax = plt.subplot()
ax.set_xticks(range(len(months)))
ax.set_xticklabels(month_names)

y_lower = [i - i * 0.1 for i in revenue]
y_upper = [i + i * 0.1 for i in revenue]

plt.plot(months, revenue)
plt.fill_between(months, y_lower, y_upper, alpha = 0.2)

plt.show()
```

### Pie Chart <a class="anchor" id="6"></a>
[Table of Contents](#0.1)

If we want to display elements of a data set as proportions of a whole, we can use a pie chart.

- Pie charts are helpful for displaying data like:
    - Different ethnicities that make up a school district
    - Different macronutrients (carbohydrates, fat, protein) that make up a meal
    - Different responses to an online poll
    
When we make pie charts in Matplotlib, we almost always want to set the axes to be equal to fix issues. To do this, we use `plt.axis('equal')`

```python
budget_data = [500, 1000, 750, 300, 100]
 
plt.pie(budget_data)
plt.axis("equal")
plt.show()
```

```python
payment_method_names = ["Card Swipe", "Cash", "Apple Pay", "Other"]
payment_method_freqs = [270, 77, 32, 11]
```

```python
plt.pie(payment_method_freqs )

plt.axis("equal")
plt.show()
```

```python
budget_data = [500, 1000, 750, 300, 100]
budget_categories = ['marketing', 'payroll', 'engineering', 'design', 'misc']
```

```python
plt.pie(budget_data, labels = budget_categories, autopct="%.1f%%")
plt.legend(budget_categories)

plt.axis("equal")
plt.show()
```

```python
plt.pie(budget_data, labels = budget_categories, autopct="%.1f%%")

plt.axis("equal")
plt.show()
```

### Histogram <a class="anchor" id="7"></a>
[Table of Contents](#0.1)

- Sometimes we want to get a feel for a large dataset with many samples beyond knowing just the basic metrics of mean, median, or standard deviation. To get more of an intuitive sense for a dataset, we can use a histogram to display all the values.


- A histogram tells us how many values in a dataset fall between different sets of numbers.


- All bins in a histogram are always the same size. The width of each bin is the distance between the minimum and maximum values of each bin.


- Histograms are best for showing the shape of a dataset. For example, you might see that values are close together, or skewed to one side. With this added intuition, we often discover other types of analysis we want to perform.

```python
import pandas as pd
```

```python
sales_times = pd.read_csv("sales_times.csv")
sales_times.head()
```

```python
pd.to_datetime(sales_times["time"])
```

```python
plt.hist(x = pd.to_datetime(sales_times["time"]), bins = 20)
plt.xticks(rotation=90);
```

### Multiple Histograms <a class="anchor" id="8"></a>
[Table of Contents](#0.1)

- **alpha :** use the keyword alpha, which can be a value between 0 and 1. This sets the transparency of the histogram. A value of 0 would make the bars entirely transparent. A value of 1 would make the bars completely opaque.


- **histtype :** use the keyword histtype with the argument 'step' to draw just the outline of a histogram


- **density = True :** This command divides the height of each column by a constant such that the total shaded area of the histogram sums to 1.

```python
sales_times2 = pd.read_csv("sales_times_s2.csv")
sales_times2.head()
```

```python
sales_times2.shape
```

```python
plt.hist(x = pd.to_datetime(sales_times["time"]), bins = 20, alpha = 0.4, density=True)
plt.xticks(rotation=90)
plt.hist(x = pd.to_datetime(sales_times2["time"]), bins = 20, alpha = 0.4, density=True)

plt.show()
```
