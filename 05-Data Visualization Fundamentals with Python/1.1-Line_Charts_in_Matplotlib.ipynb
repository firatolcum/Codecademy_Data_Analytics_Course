## Data Visualization Fundamentals

### plt.subplots_adjust()
- left : The left-side margin, with a default of 0.125.
- right : the right-side margin, with a default of 0.9.
- bottom : the bottom margin, with a default of 0.1.
- top : the top margin, with a default of 0.9
- wspace : the horizontal space between adjacent subplots, with a default of 0.2
- hspace : the vertical space between adjacent subplots, with a default of 0.2

```python
import matplotlib.pyplot as plt
```

```python
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplot(1, 2, 2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplots_adjust(wspace=0.5)

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplot(1, 2, 2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplots_adjust(bottom = 0.4)

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplot(1, 2, 2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplots_adjust(top = 0.3)

plt.show()
```

```python
plt.subplot(1, 2, 1)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplot(1, 2, 2)
plt.plot([-2, -1, 0, 1, 2], [4, 1, 0, 1, 4])

plt.subplots_adjust(left = 0.3)

plt.show()
```

### plt.legend()

```python
months = range(12)
hyrule = [63, 65, 68, 70, 72, 72, 73, 74, 71, 70, 68, 64]
kakariko = [52, 52, 53, 68, 73, 74, 74, 76, 71, 62, 58, 54]
gerudo = [98, 99, 99, 100, 99, 100, 98, 101, 101, 97, 98, 99]

plt.plot(months, hyrule)
plt.plot(months, kakariko)
plt.plot(months, gerudo)

plt.legend(labels = legend_labels, loc = 8)
plt.show()
```

### Modify Ticks
ax is an axes object, and it lets us modify the axes belonging to a specific subplot. Even if we only have one subplot, when we want to modify the ticks, we will need to start by calling either ax = plt.subplot(1, 1, 1) or ax = plt.subplot() in order to get our axes object.

```python
month_names = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep","Oct", "Nov", "Dec"]
```

```python
months = range(12)
conversion = [0.05, 0.08, 0.18, 0.28, 0.4, 0.66, 0.74, 0.78, 0.8, 0.81, 0.85, 0.85]
```

```python
ax = plt.subplot()
ax.set_xticks(months)
ax.set_xticklabels(month_names)
ax.set_yticks([0.10, 0.25, 0.5, 0.75])
ax.set_yticklabels(["10%", "25%", "50%", "75%"])

plt.plot(months, conversion)
plt.xlabel("Months")
plt.ylabel("Conversion")

plt.show()
```

```python
ax = plt.subplot()

plt.plot([0, 1, 2, 3, 4], [0, 1, 4, 9, 16])
plt.plot([0, 1, 2, 3, 4], [0, 1, 8, 27, 64])

ax.set_xticks([1, 2, 4])
plt.show()
```

```python
ax = plt.subplot()
plt.plot([1, 3, 3.5], [0.1, 0.6, 0.8], 'o')

ax.set_yticks([0.1, 0.6, 0.8])
ax.set_yticklabels(['10%', '60%', '80%'])

plt.show()
```

### Figures
In order to be sure that you dont have any stray lines, you can use the command `plt.close('all')` to clear all existing plots before you plot a new one.

Sometimes, we would rather have two separate figures. We can use the command `plt.figure()` to create new figures and size them how we want. 

Once weve created a figure, we might want to save it so that we can use it in a presentation or a website. We can use the command `plt.savefig()` to save out to many different file formats, such as png, svg, or pdf.

```python
word_length = [8, 11, 12, 11, 13, 12, 9, 9, 7, 9]
power_generated = [753.9, 768.8, 780.1, 763.7, 788.5, 782, 787.2, 806.4, 806.2, 798.9]
years = [2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009]
```

```python
x = [1 ,2 ,3 , 4 ,5 , 6, 7 ,8, 9, 0]
y1= [3, 6, 9, 12, 15, 18, 21, 24, 27, 30]
y2 = [5, 10, 15, 20, 25, 30, 35, 40, 45, 50]
```

```python
plt.plot(x, y1, c="pink", marker="o")
plt.plot(x, y2, c="gray", marker = "o")
plt.title("Two Lines on One Graph")
plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")
plt.legend(["X-y1", "X-y2"], loc=4)
```

```python
plt.subplot(1,2,1)
plt.plot(x, y1, c="pink", marker="o")
plt.subplot(1,2,2)
plt.plot(x, y2, c="gray", marker = "o")
plt.title("Two Lines on One Graph")
plt.xlabel("Amazing X-axis")
plt.ylabel("Incredible Y-axis")
plt.legend(["X-y1", "X-y2"], loc=4)
plt.tight_layout()
```
